```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defining custom autopct function for formatting percentage values
def custom_autopct(value):
    return f'{value:.2f}%'
```

```python
#data
total_turnover = 1907075789
percentage = {'UK': 32, 'Canada': 2, 'India': 2.3, 'Russia': 6, 'China': 1.70, 'Australia':8, 'USA': 20, 'France': 28}
country_turnover = {country: total_turnover * (percent / 100) for country, percent in percentage.items()}
# Print the turnover of each country
for country, turnover in country_turnover.items():
    print(f'Turnover in {country}: ${turnover:,.2f}')
```

```python
#calculating the turnover of each country using the percentage and total turnover of oct 2023
import matplotlib.pyplot as plt
#data
total_turnover = 2118077689
percentage = {'UK': 32, 'Canada': 2, 'India': 2.3, 'Russia': 6, 'China': 1.70, 'Australia':8, 'USA': 20, 'France': 28}
country_turnover = {country: total_turnover * (percent / 100) for country, percent in percentage.items()}
#Plotting
fig, ax = plt.subplots(figsize=(12, 8))
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray']
bars = ax.bar(country_turnover.keys(), country_turnover.values(), color=colors)
#Adding data labels
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f"${yval:,.2f}", ha='center', va='bottom', fontsize=10, color='black')
#Adding title and labels
ax.set_title('Turnover by Country - Oct 2023', fontsize=16)
ax.set_xlabel('Country', fontsize=14)
ax.set_ylabel('Turnover (in USD)', fontsize=14)
#Adding grid lines
ax.grid(axis='y', linestyle='--', alpha=0.7)
#Display the plot
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

### EXPENSE OF EACH COUNTRY

```python
#Total expense from the company
total_expense = 9001585
country_percentages = {'UK': 32, 'Canada': 2, 'India': 2.3, 'Russia': 6, 'China': 1.70, 'Australia':8, 'USA': 20, 'France': 28}
#Creating an empty dictionary to store country expenses
country_expenses = {}
#Calculate expenses for each country based on the provided percentages
for country, percentage in country_percentages.items():
    # Calculate the expense for the current country and round off to two decimal places
    expense_for_country = round((percentage / 100) * total_expense, 2)
    # Store the rounded expense in the dictionary
    country_expenses[country] = expense_for_country
#Plotting
fig, ax = plt.subplots(figsize=(12, 8))
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray']

bars = ax.bar(country_expenses.keys(), country_expenses.values(), color=colors)
#Adding data labels
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f"${yval:,.2f}", ha='center', va='bottom', fontsize=10, color='black')
#Adding title and labels
ax.set_title('Expenses by Country', fontsize=16)
ax.set_xlabel('Country', fontsize=14)
ax.set_ylabel('Expense (in USD)', fontsize=14)
#Adding grid lines
ax.grid(axis='y', linestyle='--', alpha=0.7)
#Display the plot
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

### EXPENSES OF UK

```python
# Given values
total_expense_uk = 2880507.20
total_expense_company = 9001585
marketing_expense_company = 1500000
collab_expense_company = 2000000
salary_expense_company = 4500000
other_expenses_company = 1001585
# Calculate the divided expense for the UK
marketing_expense_uk = (marketing_expense_company * total_expense_uk)/total_expense_company
collab_expense_uk = (collab_expense_company * total_expense_uk) / total_expense_company
salary_expense_uk = (salary_expense_company * total_expense_uk) / total_expense_company
other_expenses_uk = (other_expenses_company * total_expense_uk) / total_expense_company
# Labels and values for the pie chart
labels = ['Marketing', 'Collab', 'Salary', 'Other']
expenses_uk = [marketing_expense_uk, collab_expense_uk, salary_expense_uk, other_expenses_uk]
# Larger size for the pie chart
fig, ax = plt.subplots(figsize=(10, 8))
colors = ['blue', 'orange', 'green', 'red']
# Create a pie chart with percentages and actual values
wedges, texts, autotexts = ax.pie(expenses_uk, labels=labels, autopct=lambda p: f'${int(p*total_expense_uk/100):,.0f} ({p:.1f}%)',
                                  startangle=90, colors=colors)
# Customize text properties
for text, autotext in zip(texts, autotexts):
    text.set_fontsize(12)
    autotext.set_fontsize(10)
# Adding title
ax.set_title('Expense Distribution for the UK')
# Display the plot
plt.show()
```
